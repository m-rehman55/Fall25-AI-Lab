```python
# Simple Calculator using Stack Logic
# Yeh program ek expression ko step by step solve karta hai
# aur normal mathematical rules (BODMAS) ko follow karta hai.

# Operator ki priority check karne wala function
def precedence(op):
    if op in ['+', '-']:
        return 1
    elif op in ['*', '/']:
        return 2
    elif op in ['^', '%']:
        return 3
    return 0

# Do numbers par ek operator apply karne wala function
def apply(a, b, op):
    if op == '+':
        return a + b
    elif op == '-':
        return a - b
    elif op == '*':
        return a * b
    elif op == '/':
        return a / b
    elif op == '%':
        return a % b
    elif op == '^':
        return a ** b

# Expression solve karne ka main function
def evaluate(exp):
    nums = []      # numbers rakhenge
    ops = []       # operators rakhenge
    i = 0

    while i < len(exp):
        ch = exp[i]

        # spaces ignore kar do
        if ch == ' ':
            i += 1
            continue

        # Agar digit mila to pura number banao (multi digit bhi ho sakta hai)
        if ch.isdigit() or ch == '.':
            num_str = ""
            while i < len(exp) and (exp[i].isdigit() or exp[i] == '.'):
                num_str += exp[i]
                i += 1
            nums.append(float(num_str))
            continue

        # Opening bracket ko stack me daal do
        elif ch == '(':
            ops.append(ch)

        # Closing bracket aate hi andar ka part solve karo
        elif ch == ')':
            while ops and ops[-1] != '(':
                b = nums.pop()
                a = nums.pop()
                op = ops.pop()
                nums.append(apply(a, b, op))
            ops.pop()   # '(' hatana na bhoolna

        # Agar operator mila (+, -, *, /, etc.)
        else:
            while ops and precedence(ops[-1]) >= precedence(ch):
                b = nums.pop()
                a = nums.pop()
                op = ops.pop()
                nums.append(apply(a, b, op))
            ops.append(ch)

        i += 1

    # Agar abhi bhi koi operator bacha hai to usko bhi solve karo
    while ops:
        b = nums.pop()
        a = nums.pop()
        op = ops.pop()
        nums.append(apply(a, b, op))

    # Final answer
    return nums[-1]


# Example run
if __name__ == "__main__":
    expr = "1+2*3*(4-5/4)-(3/5)+2^3"
    print("Expression:", expr)
    ans = evaluate(expr)
    print("Result:", ans)
```
